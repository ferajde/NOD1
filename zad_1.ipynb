```python
# stworzenie tabeli przestawnej (pivot table): średnia wagi dla każdego kraju i grupy wiekowej
df_pivot = df.pivot_table(values="weight_combined", index="Country", columns="agegroup",
                          aggfunc="mean", margins=True, fill_value=0)
df_pivot
```

```python
# wyświetlenie indeksów i kolumn tabeli przestawnej
print("Indeksy:", df_pivot.index.tolist())
print("Kolumny:", df_pivot.columns.tolist())
```

```python
# tabela przestawna z indeksem złożonym (Country + agegroup)
df_pivot_multi = df.pivot_table(values="weight_combined", index=["Country", "agegroup"],
                                columns="Srvyr", aggfunc="mean", fill_value=0)
# wyświetlenie indeksu złożonego
print(df_pivot_multi.index)
```

```python
# łączenie ramek danych za pomocą merge i concat
# podział ramki danych na dwie części
df_part1 = df.iloc[:len(df)//2, :]
df_part2 = df.iloc[len(df)//2:, :]
print("Część 1:", df_part1.shape)
print("Część 2:", df_part2.shape)
```

```python
# concat: połączenie ramek danych po wierszach (axis=0)
df_concat = pd.concat([df_part1, df_part2], axis=0)
print("Po concat:", df_concat.shape)
df_concat.head()
```

```python
# merge: łączenie dwóch ramek danych po wspólnej kolumnie
df_left = df[["SubjectNumber", "Country", "agegroup", "weight_combined"]].head(100)
df_right = df[["SubjectNumber", "Srvyr", "FinalOutcome"]].head(150)
df_merged = pd.merge(df_left, df_right, on="SubjectNumber", how="inner")
print("Wynik merge:", df_merged.shape)
df_merged.head()
```

```python
# dodawanie nowej kolumny za pomocą operacji matematycznych
df["weight_x2"] = df["weight_combined"] * 2
df[["SubjectNumber", "weight_combined", "weight_x2"]].head()
```

```python
# dodawanie nowej kolumny za pomocą funkcji lambda
# przykład: oznaczenie respondentów z grupy wiekowej 3 jako senior
df["age_label"] = df["agegroup"].apply(lambda x: "senior" if x == 3 else ("middle" if x == 2 else "young"))
df[["SubjectNumber", "agegroup", "age_label"]].head(10)
```

```python
# praca z dużymi plikami przy użyciu argumentu chunksize
# wczytanie pliku CSV porcjami po 1000 wierszy i wyświetlenie 5 wierszy każdej porcji
for i, chunk in enumerate(pd.read_csv("IHME_ORB_C19HSDS_2020_Y2020M12D03.CSV", chunksize=1000)):
    print(f"Porcja {i+1}: {chunk.shape}")
    display(chunk.head(2))
    if i >= 2:  # wyświetlamy tylko pierwsze 3 porcje dla przykładu
        break
```
